# Column Details

```python
corr = data.corrwith(data['Class']).reset_index()

corr.columns = ['Index','Correlations']
corr = corr.set_index('Index')
corr = corr.sort_values(by=['Correlations'], ascending = False)

plt.figure(figsize=(8, 15))
fig = sns.heatmap(corr, annot=True, fmt="g", cmap='Set3', linewidths=0.4, linecolor='black')

plt.title("Correlation of Variables with Class", fontsize=20)
plt.show()
```

# Log transformation on all columns and perform Y vs X

```python
col_name = data.columns
col_name = col_name.drop('Class')
col_name
```

```python
data2 = pd.DataFrame()
for col in col_name:
    data2[col] = np.where(data[col]>0, np.log(data[col]), 0)
```

```python
corr2 = data2.corrwith(data['Class']).reset_index()

corr2.columns = ['Index','Correlations']
corr2 = corr2.set_index('Index')
corr2 = corr2.sort_values(by=['Correlations'], ascending = False)
```

```python
plt.figure(figsize=(8, 15))
fig = sns.heatmap(corr2, annot=True, fmt="g", cmap='Set3', linewidths=0.4, linecolor='black')

plt.title("Correlation of Variables with Class", fontsize=20)
plt.show()
```
